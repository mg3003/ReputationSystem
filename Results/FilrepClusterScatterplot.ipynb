```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\Lenovo\\Desktop\\Reputation System\\final\\filrepalgoresults.csv")
```

```python
df.columns
```

```python
plt.scatter( df['account'],df['clusters']*100, color='blue', label='Cluster Score')
plt.scatter(df['account'],df['filrepscore'], color='red', label='Filrep Score')

# Adding labels and title
plt.xlabel('Account')
plt.ylabel('Reputation Score')
plt.title('Comparison of Filrep and Cluster Scores')
plt.legend()
plt.grid()
plt.xticks(rotation=90)
plt.legend(loc='lower right')



# Display plot
plt.show()
```

```python
average_refund = df1['refund'].mean()
print("Average refund:", average_refund)
```

```python
# Count the occurrences of each success value
success_counts = df1['success'].value_counts()

# Create a bar chart
plt.bar(success_counts.index.astype(str), success_counts.values, color='green')

# Add labels and title
plt.xlabel('Success')
plt.ylabel('Count')
plt.title('Count of Success Values')
plt.grid()
# Show the plot
plt.show()

```

```python
exit_code_counts = df1['exitCode'].value_counts()

# Create a bar chart
plt.bar(exit_code_counts.index, exit_code_counts.values, color='green')

# Add labels and title
plt.xlabel('Exit Code')
plt.ylabel('Count')
plt.title('Count of Exit Codes')
plt.grid()
# Show the plot
plt.show()
```
